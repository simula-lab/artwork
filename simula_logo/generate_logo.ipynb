```python
import math
from penrose.penrose import PenroseP3, BtileL, psi
from IPython.display import SVG
```

```python
# A simple example starting with a BL tile
scale = 100
```

```python
ngen = 3
```

```python
config = {'width': '100%', 'height': '100%',
          'stroke-colour': '#000',
          'base-stroke-width': 0.02,
          'margin': 1.05,
          'tile-opacity': 0,
          'random-tile-colours': False,
          'Stile-colour': '#fff',
          'Ltile-colour': '#fff',
          'Aarc-colour': '#f00',
          'Carc-colour': '#00f',
          'draw-tiles': True,
          'draw-arcs': False,
          'reflect-x': True,
          'draw-rhombuses': True,
          'rotate': math.pi/2,
          'flip-y': False, 
          'flip-x': False,
         }
```

```python
tiling = PenroseP3(scale, ngen, config)
```

```python
theta = 2*math.pi / 5
rot = math.cos(theta) + 1j*math.sin(theta)
xOffset = -50
```

```python
A = -scale/2 + 0j + xOffset
B = scale/2 * rot + xOffset
C = scale/2 / psi + 0j + xOffset
print(A)
print(B)
print(C)
```

```python
tiling.set_initial_tiles([BtileL(A, B, C)])
tiling.make_tiling()
tiling.elements = tiling.elements[2:8] + tiling.elements[10:14]
svg = tiling.make_svg()
```
